```python
def convertRange(x):
    try:
        # Attempt to split the string if it's a range
        temp = x.split('-')
        if len(temp) == 2:
            return (float(temp[0]) + float(temp[1]))/2
    except AttributeError:
        # If it's not a string (e.g., a float), return it as is
        pass
    try:
        # Try to convert to float if possible
        return float(x)
    except (ValueError, TypeError):
        # If conversion fails, return None
        return None
```

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)

        st = np.std(subdf.price_per_sqft)

        # Corrected the typo here: price_per_sqt to price_per_sqft
        gen_df = subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft <= (m+st))]
        df_output = pd.concat([df_output,gen_df],ignore_index=True)
    return df_output
df = remove_outliers_sqft(df)
df.describe()
```

```python
def bhk_outlier_remover(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }

        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    #Corrected indentation here to align with the function definition
    return df.drop(exclude_indices,axis='index')
```

```python
df=bhk_outlier_remover(df)
```

```python
x=df.drop(columns=['price'])
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

Apply Linear Regression

```python
column_trans = make_column_transformer((OneHotEncoder(sparse_output=False),['location']),
                                       remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lr = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred_lr)
```

Applying Lasso

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lasso = pipe.predict(x_test)
r2_score(y_test,y_pred_lasso)
```

Applying Ridge

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_ridge = pipe.predict(x_test)
r2_score(y_test,y_pred_ridge)
```

```python
print("No Regularization:",r2_score(y_test,y_pred_lr))
print("Lasso:",r2_score(y_test,y_pred_lasso))
print("Ridge:",r2_score(y_test,y_pred_ridge))
```
